```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Bike_Sales_Outlier_Lab.xlsx')

print(df['Country'])
print(df['Revenue'])

plt.figure(figsize=(20,10))
plt.bar(df['Country'],df['Revenue'],color='red')
plt.xticks(rotation=90,ha='right')
plt.title('Ventas por pais',fontsize=12)
plt.xlabel('Revenue',fontsize=12)
plt.ylabel('Country',fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

duracion_llamadas= [2,3,3,4,4,5,5,5,6,6,7,7,8,9,9,10,11,12,12,13,14]

plt.hist(duracion_llamadas,bins=5,color='blue',edgecolor='black',alpha=0.7)
xmin,xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = np.exp(-0.5 * ((x - xmin) / xmin)**2) / (xmax * np.sqrt(2 * np.pi))

plt.plot(x , p ,len(duracion_llamadas)* (xmax -xmin) /5, color='red' ,linewidth=2)
plt.title('distribucion de la duracion de llamadas')
plt.xlabel('duracion')
plt.ylabel('frecuencia')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

datos = {'Clase': ['Clase A']*10 + ['Clase B']*10,
         'Puntaje': [65, 70, 75, 80, 85, 90, 85, 80, 75, 70,
                     60, 65, 70, 75, 80, 85, 90, 95, 100, 105]}


df= pd.DataFrame(datos)

sns.boxplot(x='Clase', y='Puntaje', data=df,palette='Set2')
plt.title('Distribucion de puntajes por clase',fontsize=16)
plt.xlabel('Clase',fontsize=12)
plt.ylabel('Puntaje',fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('/content/Bike_Sales_Outlier_Lab.xlsx')



plt.figure(figsize=(20,10))
plt.scatter(df['Customer_Age'],df['Order_Quantity'], color='red', s=100,edgecolors='black',alpha=0.7)

m,b=  np.polyfit(df['Customer_Age'],df['Order_Quantity'],1)

plt.plot(df['Customer_Age'],df['Customer_Age']+b, color='red' , label='correlacion')
plt.title('edad vs cantidad',fontsize=12)
plt.xlabel('edad cliente',fontsize=12)
plt.ylabel('cantidad',fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```
